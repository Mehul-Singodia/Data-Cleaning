```python
import pandas as pd
import numpy as np
```

```python
# Loading the data
data = pd.read_excel('Student_Scorecard.xlsx')
data = data.sort_values('Student ID').reset_index(drop = True)
data
```

```python
# Inputing pin fields
pin_fields = input('Enter the pin field (seperated by ,): ')
```

```python
# Transforimg the input into list contaning pin_fileds

pin_fields = pin_fields.split(',')

for i in range(len(pin_fields)):
    pin_fields[i] = pin_fields[i].strip()

print(pin_fields)
```

```python
# Droping the dublicates and keeping the first occurance
unique_data = data.drop_duplicates(subset = pin_fields, keep = 'first', ignore_index = False)  # ignore_index will make the index continous


# rows which are dropped                                                    # first unit is not removed
all_rem_rows = data[data.duplicated(subset = pin_fields, keep = 'first')]   # exactly all rows  [might contain exact similar rows as well]
unq_rem_rows = all_rem_rows.drop_duplicates().reset_index(drop =True)       # distinct removed rows

# Count of duplicates of each duplicate row

grp_dup_data = all_rem_rows.groupby(pin_fields)

dup_counts = grp_dup_data.size().reset_index(name='Count')                  # including the unique one
dup_counts['Count'] = dup_counts['Count'] + 1

complete_dup_counts = data[data[pin_fields].apply(tuple, axis=1).isin(dup_counts[list(pin_fields)].apply(tuple, axis=1))]

data = unique_data
data
```
